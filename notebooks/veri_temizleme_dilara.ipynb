```python
import pandas as pd 

customers = pd.read_excel(r"C:\Users\dilara\Desktop\Customers.xlsx")
customers.head()
```

```python
customers.info()
```

```python
customers['Region'] = customers['Region'].fillna('Bilinmiyor')
customers['Fax'] = customers['Fax'].fillna('Yok')
customers['PostalCode'] = customers['PostalCode'].fillna('Bilinmiyor')
```

```python
customers.info()
```

```python
customers['Country'] = customers['Country'].astype('category')
customers['City'] = customers['City'].astype('category')
```

```python
customers['Phone_Clean'] = customers['Phone'].str.replace(r'[\(\)\s\-]', '', regex=True)
```

```python
duplicate_count = customers.duplicated().sum()
print(f"Tekrar eden satır sayısı: {duplicate_count}")
```

```python
customers.head()
```

```python
clean_customers = customers.copy()
clean_customers.to_csv("clean_customers.csv", index=False)
print("Veri 'clean_customers' değişkenine atandı ve 'clean_customers.csv' olarak kaydedildi.")
```

```python
ordersdetails = pd.read_excel(r"C:\Users\dilara\Desktop\OrdersDetails.xlsx")
ordersdetails.head()
```

```python
ordersdetails.info()
```

```python
ordersdetails['UnitPrice'] = ordersdetails['UnitPrice'].str.replace(',', '.').astype(float)
```

```python
# İndirimi de hesaba katarak toplam tutarı hesaplama
ordersdetails['TotalAmount'] = (ordersdetails['UnitPrice'] * ordersdetails['Quantity']) * (1 - ordersdetails['Discount'])
```

```python
print(f"Tekrar eden satır sayısı: {ordersdetails.duplicated().sum()}")
```

```python
ordersdetails.info()
```

```python
clean_ordersdetails = ordersdetails.copy()
clean_ordersdetails.to_csv("clean_ordersdetails.csv", index=False)
print("Veri 'clean_ordersdetails' değişkenine atandı ve 'clean_ordersdetails.csv' olarak kaydedildi.")
```

```python
orders = pd.read_excel(r"C:\Users\dilara\Desktop\Orders.xlsx")
orders.head()
```

```python
import os
import shutil

# Hedef klasörü oluştur (yoksa)
target_dir = "../data/processed/"
os.makedirs(target_dir, exist_ok=True)

# Taşınacak dosyaların listesi
files_to_move = [
    "clean_customers.csv",
    "clean_orders.csv",
    "clean_ordersdetails.csv",
    "clean_products.csv",
    "clean_categories.csv"
]

for file in files_to_move:
    if os.path.exists(file):
        # Dosyayı notebooks klasöründen data/processed klasörüne taşı
        shutil.move(file, os.path.join(target_dir, file))
        print(f"{file} başarıyla {target_dir} altına taşındı.")
    else:
        print(f"Uyarı: {file} bulunamadı.")
```
